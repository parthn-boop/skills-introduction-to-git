```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
documents = [
    "Machine learning is a subset of artificial intelligence.",
    "Deep learning uses neural networks.",
    "Python is popular for data science.",
    "Pandas is used for data analysis.",
    "NumPy performs numerical computations.",
    "TensorFlow is a deep learning framework.",
    "PyTorch is widely used for neural networks.",
    "Natural language processing deals with text data.",
    "Transformers are powerful NLP models.",
    "BERT is based on the transformer architecture.",
    "Random forests are ensemble learning methods.",
    "Decision trees are easy to understand.",
    "K-means is a clustering algorithm.",
    "Linear regression predicts continuous values.",
    "Logistic regression is used for classification.",
    "CNNs are used for image processing.",
    "RNNs process sequential data.",
    "Cosine similarity measures vector similarity.",
    "Embeddings capture semantic meaning.",
    "Semantic search retrieves relevant documents."
]
```

```python
doc_embeddings = model.encode(documents)
```

```python
sentence1 = "Python is popular for data science."
sentence2 = "Pandas is used for data analysis."

embedding1 = model.encode([sentence1])
embedding2 = model.encode([sentence2])

similarity = cosine_similarity(embedding1, embedding2)

print("Cosine Similarity:", similarity[0][0])
```

```python
def semantic_search(query, top_k=3):

    # Convert query to embedding
    query_embedding = model.encode([query])

    # Calculate similarities
    similarities = cosine_similarity(query_embedding, doc_embeddings)[0]

    # Get indices of top-k documents
    top_indices = similarities.argsort()[-top_k:][::-1]

    print("Query:", query)
    print()

    for idx in top_indices:
        print("Similarity Score:", round(similarities[idx], 4))
        print("Document:", documents[idx])
        print()
```

```python
semantic_search("Algorithms for grouping similar data")
```

```python
semantic_search( "Transformers are powerful NLP models.")
```

```python
semantic_search( "Machine learning is a subset of artificial intelligence.")
```
